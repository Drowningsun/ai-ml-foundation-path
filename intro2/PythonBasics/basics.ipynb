*PATTERN PRINTING*

```python
row = 10

for i in range(row+1):
    for j in range((row - i)):
        print(" ",end = "")
    for k in range(i):
        if k <= 0:
            print("*",end="")
        else:
            print("*"*2,end="")
    for l in range((row - i)):
        print(" ",end = "")
    print()
```

```python
import datetime

# Current date
today = datetime.date.today()
print(type(today))
print(f"Today's date: {today}")

# Date 100 days from today
future_date = today + datetime.timedelta(days=100)
print(f"Date 100 days from today: {future_date}")

# Day of the week for a given date
given_date = datetime.date(2022, 1, 1)
print(f"Day of the week for 2022-01-01: {given_date.strftime('%A')}")
```

```python
import csv

def read_csv_as_dicts(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        return list(reader)
    
data = read_csv_as_dicts('industry.csv')
print(data)
```
