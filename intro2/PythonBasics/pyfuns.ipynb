```python
memo={}

def febanochicci(n):
    if n in memo:
        return memo[n]
    
    if n <= 1 :
        memo[n] = n
    else:
        memo[n] = febanochicci(n-1) + febanochicci(n-2)
    
    return memo[n]

print(febanochicci(5))
print(febanochicci(10))
print(memo)
```

```python
def add_to_dect(a,b={}):
    b[a] = a**2
    return b

print(add_to_dect(13,{11:121,5:25}))
```

```python
def integer_value_filtering(**kwargs):
    filtered_dict = {key:value for key,value in kwargs.items() if type(value) == int}

    return filtered_dict

integer_value_filtering(name = "hari", age = 55, marks = 90, grade = "A")
```
