#### dictonaries

```python
dict1 = {x:x**2 for x in range(1,10+1)}
print(dict1)
```

```python
print(dict1[5])
print(dict1.keys())``
```

```python
dict1[11] = 121
print(dict1)
```

```python
for key,value in dict1.items():
    print(f"{key}:{value}")
```

```python
dict2 = {x:x**3 for x in range(1,10+1)}
print(dict2)
```

```python
dict3 = {x:x**2 for x in range(1,5+1)}
dict4 = {x:x**2 for x in range(6,10+1)}

print(dict3)
print(dict4)

dict_merged = {**dict3, **dict4}
print(dict_merged)
```

```python
student_dict = {
    "name":"jimmy",
    "age":15,
    "grade":{"maths":85, "science":90, "english":88}
}
print(student_dict)
```

```python
list_dict = {x:[mul*x for mul in range(1,5+1)] for x in range(1,5+1)}
print(list_dict)
```

```python
tuple_dict = {i: (i, i**2) for i in range(1, 6)}
print(tuple_dict)
```

```python
d = {i: i**2 for i in range(1, 6)}
list_of_tuples = list(d.items())
print(list_of_tuples)
```

```python
sqr_dict = {x:x**2 for x in range(1,10+1)}
even_square_dict = {key:value for key,value in sqr_dict.items() if key%2 == 0}
print(sqr_dict)
print(even_square_dict)
```

```python
from collections import defaultdict

default_dict = defaultdict(list)
default_dict['a'].append(1)
default_dict['a'].append(2)
default_dict['b'].append(3)
print(default_dict)
```

```python
def character_frq(strng):
    freq_dict = {}
    for c in strng:
        if c in freq_dict:
            freq_dict[c] += 1
        else:
            freq_dict[c] = 1
    return freq_dict


character_frq("wassap my nigga")
```
